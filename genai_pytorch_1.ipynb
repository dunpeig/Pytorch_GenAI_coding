```python
import torch 
import torch.nn as nn 
import torchvision 
import torchvision.transforms as T 

torch.manual_seed(42)
transform = T.Compose([
    T.ToTensor(),
    T.Normalize([0.5], [0.5])
])
```

```python
train_set = torchvision.datasets.FashionMNIST(
    root=".",
    train=True,
    download=True, 
    transform=transform
)
test_set = torchvision.datasets.FashionMNIST(root=".",
                                             train=False, download=True, transform=transform)
```

```python
print(train_set[0])
```

```python
# got labels here https://github.com/pranay414/Fashion-MNIST-Pytorch 
text_labels = ['t-shirt', 'trouser', 'pullover', 'dress', 'coat',
                'sandal', 'shirt', 'sneaker', 'bag', 'ankle boot']
```

```python
binary_train_set = [x for x in train_set if x[1] in [0,9]]
# image tensor x[0] of shape 1, 28, 28
# x[1] the image label 0 - 9 
binary_test_set = [x for x in test_set if x[1] in [0,9]]

# parellel data loading with batch size and shuffleing the samples 
batch_size = 64 
binary_train_loader = torch.utils.data.DataLoader(
    binary_train_set, 
    batch_size=batch_size,
    shuffle=True
)
binary_test_loader = torch.utils.data.DataLoader(
    binary_test_set,
    batch_size=batch_size, shuffle=True
)
# what does torch.utils.data.DataLoader do?
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

binary_model = nn.Sequential(
    # flatterns 28 by 28 imaghes (784 pixels) to 258 neurons 
    nn.Linear(28*28, 258),
    # relu activation function that keeps only positive values
    nn.ReLU(),
    # hidden layer 
    nn.Linear(258, 128),
    nn.Linear(128, 32),
    nn.ReLU(),
    nn.Linear(32,1),
    # randonly dropped out 25 percent of neurons 
    nn.Dropout(p=0.25),
    nn.Sigmoid()
).to(device)
```

```python
lr = 0.001 
# amam optimizer , pass in the parameters of the model to optimize
optimizer = torch.optim.Adam(binary_model.parameters(), lr=lr)
# what is BCELoss?
loss_fn = nn.BCELoss()
```

```python
for i in range(50):
    tloss = 0.0 
    # enumerate the binary train loader for each imgs and label
    for n,(imgs, labels) in enumerate(binary_train_loader):
        # batch size of 64, shape 1, 28, 28
        # flatterns to one dimension vector of 784 , -1 infers the batch size 
        # so shape becomes 64, 784
        imgs = imgs.reshape(-1, 28*28)
        imgs = imgs.to(device)
        # comvert to binary , float type tensor 
        labels = torch.FloatTensor(\
            [x if x ==0 else 1 for x in labels])
        # also infer the batch size and one, the binary label 
        labels = labels.reshape(-1,1).to(device)
        preds = binary_model(imgs)
        # BCE is the binary cross entropy loss 
        # BCE expects the input go through a sigmoid function, which is 
        # specified in the model above 
        loss = loss_fn(preds, labels)
        # before computing the new gradients , must clear old ones 
        # resets gradients to zero so each backward pass starts fresh 
        optimizer.zero_grad()
        # computes the gradients of the loss with respct to 
        # model parameters 
        # pytorch automatically tracks all operations during the forward pass
        # and uses that information to compute gradients during backward pass
        loss.backward()
        # updates the model parameters based on the computed gradients, how Adam modifies the weights
        optimizer.step()
        tloss += loss 
    # divide by the batches to normalize the loss 
    tloss = tloss / n 
    print(f"at epoch {i}, loss is {tloss}")
```

# When your model has multiple layers, each layer transforms the data in sequence 
# and during training, PyTorch automatically computes how each weight in each layer contributed to the final loss.

# This works thanks to automatic differentiation (Autograd)  PyTorchs engine that builds a computational graph as you 
# run your forward pass.

```python
import numpy as np 

results = []
for imgs, labels in binary_test_loader:
    imgs = imgs.reshape(-1, 28*28).to(device)
    labels = torch.FloatTensor(\
        [x if x == 0 else 1 for x in labels]
        )
    labels = (labels).reshape(-1,1).to(device)
    preds = binary_model(imgs)
    pred10 = torch.where(preds>0.5,1,0)
    correct = (pred10==labels)
    results.append(correct.detach().cpu()\
                   .numpy().mean())
accuracy = np.array(results).mean()
print(f"the accuracy of the prediction is {accuracy}")
```

```python
torch.manual_seed(42)
# split the train set into train and validation sets
train_set, val_set = torch.utils.data.random_split(\
    train_set, [50000,10000])
train_loader = torch.utils.data.DataLoader(
    train_set,
    batch_size=batch_size,
    shuffle=True
)
val_loader = torch.utils.data.DataLoader(
    val_set, 
    batch_size=batch_size,
    shuffle=True
)
test_loader=torch.utils.data.DataLoader(
    test_set,
    batch_size=batch_size,
    shuffle=True
)
```
